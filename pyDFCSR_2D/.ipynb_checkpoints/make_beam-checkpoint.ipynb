```python
from distgen import Generator
import numpy as np
```

```python
filename = './input/dipole_init_beam.yaml'
gen = Generator(filename)
gen.run()
pg = gen.particles
gamma = pg.gamma
delta = (gamma - np.mean(gamma))/np.mean(gamma)
particles = np.vstack((pg.x, pg.xp, pg.y, pg.yp, pg.z, delta)).T
charge = pg['charge']
init_energy = np.mean(pg['energy'])
```

```python
np.savetxt('./input_beam/dipole_beam.dat', particles)
```

```python
import h5py
import matplotlib.pyplot as plt
```

```python
with h5py.File("input_beam/preB2LE1_lucretiaCSR", "r") as f:
    x = np.array(f['x'])
    xp = np.array(f['xp'])
    y = np.array(f['y'])
    yp = np.array(f['yp'])
    z = np.array(f['z'])
    E = np.array(f['G'])
    E0 = np.mean(E)
    delta = (E -E0)/E0
```

```python
x -= np.mean(x)
z -= np.mean(z)
y -= np.mean(y)
```

```python
particles = np.vstack((x, xp, y, yp, z, delta)).T
```

```python
fig =plt.hist2d(z, x, bins = 200)
```

```python
ind = (x > -0.0002)&(z >-0.00005) & (z < 0.00001)
xcut = x[ind]
zcut = z[ind]
xpcut = xp[ind]
ypcut = yp[ind]
Ecut = E[ind]
E0 = np.mean(Ecut)
dcut = (Ecut -E0)/E0
```

```python
fig =plt.hist2d(z, delta, bins = 200)
```

```python
fig =plt.hist2d(zcut, xcut, bins = 200)
```

```python
particles = np.vstack((xcut, xpcut, ypcut, ypcut, zcut, dcut)).T
```
